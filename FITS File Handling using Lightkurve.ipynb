```python
def fits_file_to_np_array(filename, flux_type="PDCSAP_FLUX"):
    KPLR = lk.lightcurvefile.KeplerLightCurveFile(filename)
    return KPLR.get_lightcurve(flux_type).flux

```

```python
import matplotlib.pyplot as plt
plt.plot(fits_file_to_np_array('kplr010000009-2009131105131_llc.fits'))
```
